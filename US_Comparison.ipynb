# Compare renewable production vs consumption for US

```python
# import dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
```

```python
# read the production data
production_df = pd.read_excel("db/original-project3-prod-data.xlsx")
production_df
```

```python
# select only renewable type 
production_df = production_df.loc[production_df["MSN"]=="REPRB"]
production_df
```

```python
# select us total renewable data
production_df = production_df.loc[production_df["StateCode"] == "US"]
production_df
```

```python
# Drop the columns we don't need
prod_df = production_df.drop(["Data_Status", "StateCode", "MSN"], axis = 1)
prod_df
```

```python
# convert columns to rows 
prod_data_df = prod_df.melt(var_name= "Year", 
        value_name="Prod Data")
pd.set_option('display.max_rows', None)
prod_data_df.head()
```

```python
prod_data_df.info()
```

```python
# change data type of the year column
prod_data_df["Year"] = prod_data_df["Year"].astype(str).astype(int)
prod_data_df.info()
```

```python
# Read the consumption data
consumption_df = pd.read_csv("db/use_all_btu.csv")
consumption_df.head()
```

```python
# Select oly total consumed energy type only
consumption_df = consumption_df.loc[consumption_df["MSN"] == "TETCB"]
consumption_df.head()
```

```python
# Select tottal US consumption
consumption_df = consumption_df.loc[consumption_df["State"] == "US"]
consumption_df
```

```python
# Drop unnecessary columns
cons_df = consumption_df.drop(["Data_Status", "State", "MSN"], axis = 1)
cons_df
```

```python
# convert rows and columns
cons_data_df = cons_df.melt(var_name= "Year", 
        value_name="Cons Data")
pd.set_option('display.max_rows', None)
cons_data_df.head()
```

```python
# see the dataframe info
cons_data_df.info()
```

```python
# convert year data type to integer

cons_data_df["Year"] = cons_data_df["Year"].astype(str).astype(int)
prod_data_df.info()
```

```python
# join prodoction and consumption dataFrames
us_comparison_df = pd.merge(prod_data_df, cons_data_df, on="Year")
us_comparison_df
```

```python
# Add a new column to label if production can meet the consumption values
us_comparison_df["Status"] = np.where(us_comparison_df["Prod Data"] > us_comparison_df["Cons Data"], 1, 0)
us_comparison_df    
```

```python
# add a new column 
us_comparison_df["Comparison"] =us_comparison_df["Prod Data"] - us_comparison_df["Cons Data"]
us_comparison_df.head()
```

```python
# save the dataframe as csv
us_comparison_df.to_csv("db/us_comparison.csv")
```
